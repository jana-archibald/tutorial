### If you want to use the normalplot function you need to run the code below.

```r
normalplot<-function(m,sd,region=0){
  x<-seq(m-(3.5)*sd,m+(3.5)*sd,length=1000)
  y<-dnorm(x,m,sd)
  plot(x,y,type="l",xlab="",ylab="", bty="n", yaxt="n")
  h <- dnorm(m,m,sd)
  z<-x[x>region[1]]
  z<-z[z<region[2]]
  polygon(c(region[1],z,region[2]),
          c(0,dnorm(z,m,sd),0),col="gray")
  abline(v=m)
  abline(h=0)}
```

Suppose X is a random variable and X~N(80,20)

```r
normalplot(80,20)
```

```r
pnorm(65,80,20)
normalplot(80,20,c(0,65))
```

The shaded region is P(X<65)

```r
qnorm(0.3,80,20)
normalplot(80,20,c(0,qnorm(0.3,80,20)))
```

The area of the shaded region is 0.30.  So P(X<69.5119897458392)=0.3

```r
normalplot(14.7,33)
```

SAT scores (out of 1600) are distributed normally with a mean of 1100 and a standard deviation of 200. 

```r
normalplot(1100,200)
```

```r
1-pnorm(1350,1100,200)
normalplot(1100,200,c(1350,2000))
```

```r
normalplot(1100,200,c(1500,2000))
1-pnorm(1500,1100,200)
```

The probability of selecting a student with a SAT score over 1500 is 0.0227501319481792

Suppose a school council awards a certificate of excellence to all students who score at least 1350 on the SAT, and suppose we pick one of the recognized students at random. 

What is the probability this students score will be at least 1500? 
(Hint try writing this as a conditional  probability)

We know that $P(A|B)=\frac{P(A and B)}{P(B)}$.

We want to compute P(more than 1500|more than 1350).

Since everyone who scores over 1500 is scoring over 1350, P(more than 1500 and more than 1350)=P(more than 1500)

That means what we need to compute is:

P(more than 1500|more than 1350)=P(more than 1500)/P(more than 1350)

```r
0.0227501319481792/0.105649773666855
```
